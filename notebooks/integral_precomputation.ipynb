# Imports and setup

```python
from src.poroelastic_kernels import p_sm, p_em, sigma_xx_em, sigma_yy_em 
from src.integration_utils import (
    integral_near_tip_cauchy,
    integral_near_tip_log,
    integral_intervals_log,
    integral_intervals_cauchy,
)

%load_ext autoreload
%autoreload 2
```

# Atkinson-Craster benchmarks

```python
num_points = 300

xi_range = np.logspace(-8, 10, num_points)
xi_mid_range = 0.5 * (xi_range[1:] + xi_range[:-1])

power_1 = 0.5
power_2 = 1.5

beta = 0.25
```

```python
p_s_near_field_integral = integral_near_tip_log(xi_mid_range, xi_range[0], power_1, p_sm)
p_s_matrix_0 = integral_intervals_log(xi_mid_range, xi_range, power_1, p_sm)
p_s_matrix_inf = integral_intervals_log(xi_mid_range, xi_range, power_2, p_sm)
```

```python
p_em_kernel = lambda x: p_em(x, False)

p_e_near_field_integral = integral_near_tip_cauchy(xi_mid_range, xi_range[0], power_1, p_em_kernel)
p_e_matrix_0 = integral_intervals_cauchy(xi_mid_range, xi_range, power_1, p_em_kernel)
p_e_matrix_inf = integral_intervals_cauchy(xi_mid_range, xi_range, power_2, p_em_kernel)
```

```python
s_em_kernel = lambda x, beta=beta: sigma_yy_em(x, beta, False)

s_e_near_field_integral = integral_near_tip_cauchy(xi_mid_range, xi_range[0], power_1, s_em_kernel)
s_e_matrix_0 = integral_intervals_cauchy(xi_mid_range, xi_range, power_1, s_em_kernel)
s_e_matrix_inf = integral_intervals_cauchy(xi_mid_range, xi_range, power_2, s_em_kernel)
```

```python
s_em_kernel = lambda x: sigma_yy_em(x, 0, False)

cauchy_near_field_integral = integral_near_tip_cauchy(xi_mid_range, xi_range[0], power_1, s_em_kernel)
cauchy_matrix_0 = integral_intervals_cauchy(xi_mid_range, xi_range, power_1, s_em_kernel)
cauchy_matrix_inf = integral_intervals_cauchy(xi_mid_range, xi_range, power_2, s_em_kernel)
```
